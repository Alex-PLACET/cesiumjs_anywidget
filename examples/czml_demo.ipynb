## Setup

```python
from cesiumjs_anywidget import CesiumWidget
import json
```

## 1. Simple Point with Label

Create a basic CZML document with a single point and label.

```python
# Create widget
widget = CesiumWidget(height="600px")
widget
```

```python
# Simple point CZML
simple_point_czml = [
    {
        "id": "document",
        "name": "Simple Point Demo",
        "version": "1.0"
    },
    {
        "id": "point1",
        "name": "Eiffel Tower",
        "position": {
            "cartographicDegrees": [2.2945, 48.8584, 100]
        },
        "point": {
            "pixelSize": 10,
            "color": {
                "rgba": [255, 0, 0, 255]
            }
        },
        "label": {
            "text": "Eiffel Tower",
            "font": "14pt sans-serif",
            "fillColor": {
                "rgba": [255, 255, 255, 255]
            },
            "outlineColor": {
                "rgba": [0, 0, 0, 255]
            },
            "outlineWidth": 2,
            "pixelOffset": {
                "cartesian2": [0, -20]
            }
        }
    }
]

widget.load_czml(simple_point_czml)
```

## 2. Multiple Points

Load multiple points representing famous landmarks.

```python
# Multiple landmarks
landmarks_czml = [
    {
        "id": "document",
        "name": "World Landmarks",
        "version": "1.0"
    },
    {
        "id": "liberty",
        "name": "Statue of Liberty",
        "position": {"cartographicDegrees": [-74.0445, 40.6892, 100]},
        "point": {"pixelSize": 12, "color": {"rgba": [0, 255, 0, 255]}},
        "label": {
            "text": "New York",
            "pixelOffset": {"cartesian2": [0, -20]}
        }
    },
    {
        "id": "opera",
        "name": "Sydney Opera House",
        "position": {"cartographicDegrees": [151.2153, -33.8568, 50]},
        "point": {"pixelSize": 12, "color": {"rgba": [0, 0, 255, 255]}},
        "label": {
            "text": "Sydney",
            "pixelOffset": {"cartesian2": [0, -20]}
        }
    }
]

widget.load_czml(landmarks_czml)
```

## 3. Path/Polyline

Create a path between points.

```python
# Path CZML
path_czml = [
    {
        "id": "document",
        "name": "Flight Path",
        "version": "1.0"
    },
    {
        "id": "flight_path",
        "name": "Paris to New York",
        "polyline": {
            "positions": {
                "cartographicDegrees": [
                    2.2945, 48.8584, 10000,    # Paris
                    -10, 52, 10000,             # Mid-Atlantic
                    -40, 50, 10000,
                    -74.0445, 40.6892, 10000   # New York
                ]
            },
            "material": {
                "solidColor": {
                    "color": {
                        "rgba": [255, 255, 0, 255]
                    }
                }
            },
            "width": 5,
            "clampToGround": False
        }
    }
]

widget.load_czml(path_czml)
```

## 4. Polygon

Create a polygon area.

```python
# Polygon CZML
polygon_czml = [
    {
        "id": "document",
        "name": "Polygon Demo",
        "version": "1.0"
    },
    {
        "id": "area1",
        "name": "Sample Area",
        "polygon": {
            "positions": {
                "cartographicDegrees": [
                    -75.0, 45.0, 0,
                    -70.0, 45.0, 0,
                    -70.0, 40.0, 0,
                    -75.0, 40.0, 0
                ]
            },
            "material": {
                "solidColor": {
                    "color": {
                        "rgba": [0, 255, 0, 100]
                    }
                }
            },
            "outline": True,
            "outlineColor": {
                "rgba": [0, 255, 0, 255]
            },
            "outlineWidth": 2
        }
    }
]

widget.load_czml(polygon_czml)
```

1. **CZML Structure**: Every CZML document must start with a document packet (id="document")
2. **Coordinates**: Use `cartographicDegrees` format: [longitude, latitude, height]
3. **Colors**: RGBA format with values 0-255
4. **Time-Dynamic**: Use `epoch` and time-series arrays for animations
5. **Multiple Entities**: Array can contain multiple entity packets
6. **Validation**: Use the [CZML Validator](https://github.com/AnalyticalGraphicsInc/czml-writer) for complex documents

For more information, see the [CZML Guide](https://github.com/AnalyticalGraphicsInc/czml-writer/wiki/CZML-Guide).
